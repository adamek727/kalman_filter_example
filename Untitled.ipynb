```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
t_start = 0
t_end = 10
dt = 0.1
sample_num = int((t_end-t_start)/dt + 1)
time_line = np.linspace(t_start, t_end, sample_num)

pose_noise_sigma = 0.1
speed_noise_sigma = 0.1
acc_noise_sigma = 0.1
```

```python
pose_x = np.sin(time_line)
pose_noise_x = pose_x + np.random.normal(0, pose_noise_sigma, sample_num)

speed_x = np.cos(time_line)
speed_noise_x = speed_x + np.random.normal(0, speed_noise_sigma, sample_num)

acc_x = -np.sin(time_line)
acc_noise_x = acc_x + np.random.normal(0, acc_noise_sigma, sample_num)

pose_y = np.cos(time_line)
pose_noise_y = pose_y + np.random.normal(0, pose_noise_sigma, sample_num)

speed_y = -np.sin(time_line)
speed_noise_y = speed_y + np.random.normal(0, speed_noise_sigma, sample_num)

acc_y = -np.cos(time_line)
acc_noise_y = acc_y + np.random.normal(0, acc_noise_sigma, sample_num)
```

```python
print('* Pose *')
plt.plot(time_line, pose_x, 'g*')
plt.plot(time_line, pose_noise_x, 'r--')
plt.show()

plt.plot(time_line, pose_y, 'g*')
plt.plot(time_line, pose_noise_y, 'r--')
plt.show()

print('* Speed *')
plt.plot(time_line, speed_x, 'b')
plt.plot(time_line, speed_noise_x, 'k--')
plt.show()

plt.plot(time_line, speed_y, 'b')
plt.plot(time_line, speed_noise_y, 'k--')
plt.show()

print('* Acc *')
plt.plot(time_line, acc_x, 'y')
plt.plot(time_line, acc_noise_x, 'g--')
plt.show()

plt.plot(time_line, acc_y, 'y')
plt.plot(time_line, acc_noise_y, 'g--')
plt.show()

print('* Trajectory *')
plt.plot(pose_x, pose_y, 'g*')
plt.plot(pose_noise_x, pose_noise_y, 'r--')
plt.show()
```

```python
class Kalman1D:
    
    def __init__(self):
        self.state = np.matrix([[0],
                                [0]], dtype = np.float)
        self.covariance = np.matrix([[1, 0],
                                     [0, 1]], dtype = np.float)
    
    def prediction_update(self, u, dt):
#         print('Prediction')
        A = self._get_model_matrix(dt)
        B = self._get_control_matrix(dt)
        Q = self._get_covariance_process_noise_matrix(dt)
        self.state = A * self.state + B *  u
        self.covariance = A * self.covariance * np.transpose(A) + Q
    
    def measurement_update(self, measurement):
#         print('Measurement')
        H = self._get_measurement_matrix()
        R = self._get_covariance_observation_noise_matrix()
        K = self.covariance * np.transpose(H) / (H * self.covariance * np.transpose(H) + R)
        self.state = self.state + (K * (measurement - H*self.state))
        self.covariance = (np.eye(2) - K*H) * self.covariance
                      
    def get_pose(self):
        return np.asscalar(self.state[0][0])
    
    def get_speed(self):
        return np.asscalar(self.state[1][0])
    
    def get_covariance(self):
        return self.covariance
    
    @staticmethod
    def _get_model_matrix(dt): # A
        return np.matrix([[1, dt],
                          [0, 1 ]], dtype = np.float)
    
    @staticmethod
    def _get_control_matrix(dt): # B
        return np.matrix([[0.5*dt*dt],
                          [dt]], dtype = np.float)
    
    @staticmethod
    def _get_measurement_matrix(): # H
        return np.matrix([[1, 0],
                          [0, 0]], dtype = np.float)
    
    @staticmethod
    def _get_covariance_process_noise_matrix(dt): # Q
        sigma_a = 1
        return np.matrix([[math.pow(dt,4)*sigma_a, math.pow(dt,3)*sigma_a],
                          [math.pow(dt,3)*sigma_a, math.pow(dt,2)*sigma_a]], dtype = np.float)
    
    @staticmethod
    def _get_covariance_observation_noise_matrix(): # R
        return np.matrix([0.5], dtype = np.float)
```

```python
filtered_pose_x = list()
filtered_speed_x = list()
covariance_x = list()
kalman_filter_x = Kalman1D()
for time, mp, ms, ma in zip(time_line, pose_noise_x, speed_noise_x, acc_noise_x):
    kalman_filter_x.prediction_update(ma, dt)
    kalman_filter_x.measurement_update( np.matrix([[mp], [ms]]) )
    filtered_pose_x.append(kalman_filter_x.get_pose())
    filtered_speed_x.append(kalman_filter_x.get_speed())
    covariance_x.append(kalman_filter_x.get_covariance())
    
filtered_pose_y = list()
filtered_speed_y = list()
kalman_filter_y = Kalman1D()
for time, mp, ms, ma in zip(time_line, pose_noise_y, speed_noise_y, acc_noise_y):
    kalman_filter_y.prediction_update(ma, dt)
    kalman_filter_y.measurement_update( np.matrix([[mp], [ms]]) )
    filtered_pose_y.append(kalman_filter_y.get_pose())
    filtered_speed_y.append(kalman_filter_y.get_speed())
    
print('* Pose *')
plt.plot(time_line, pose_x, 'g*')
plt.plot(time_line, pose_noise_x, 'r--')
plt.plot(time_line, filtered_pose_x, 'b*')
# plt.plot(time_line, filtered_speed, 'k--')
plt.show()

plt.plot(pose_x, pose_y, 'g*')
plt.plot(pose_noise_x, pose_noise_y, 'r--')
plt.plot(filtered_pose_x, filtered_pose_y, 'b--')
```

```python
cov_11 = list()
cov_12 = list()
cov_21 = list()
cov_22 = list()
for c in covariance_x:
    cov_11.append(np.asscalar(c[0,0]))
    cov_12.append(np.asscalar(c[0,1]))
    cov_21.append(np.asscalar(c[1,0]))
    cov_22.append(np.asscalar(c[1,1]))
    
print('* Covariance X * ')
plt.plot(time_line, cov_11, 'r')
plt.plot(time_line, cov_12, 'g')
plt.plot(time_line, cov_21, 'b')
plt.plot(time_line, cov_22, 'k')
plt.show()
```

```python
noise_sigma = 0.1

unit_step = np.zeros((sample_num), dtype = np.float)
unit_step[int(sample_num/2):] = 1
unit_step_noise = unit_step + np.random.normal(0, noise_sigma, sample_num)

dirac = np.zeros((sample_num), dtype = np.float)
dirac[int(sample_num/2)] = 1
dirac_noise = dirac + np.random.normal(0, noise_sigma, sample_num)

dirac_diff = np.zeros((sample_num), dtype = np.float)
dirac_diff[int(sample_num/2)] = 1 / dt
dirac_diff[int(sample_num/2)+1] = -1 / dt
dirac_diff_noise = dirac_diff + np.random.normal(0, noise_sigma, sample_num)
```

```python
plt.plot(time_line, unit_step, 'r')
plt.plot(time_line, dirac, 'g*')
plt.plot(time_line, dirac_diff, 'b--')
plt.show()

plt.plot(time_line, unit_step_noise, 'r')
plt.plot(time_line, dirac_noise, 'g*')
plt.plot(time_line, dirac_diff_noise, 'b--')
plt.show()
```

```python
filtered_pose = list()
filtered_speed = list()
covariance = list()
kalman_filter = Kalman1D()
for time, mp, ms, ma in zip(time_line, unit_step, dirac, dirac_diff):
    kalman_filter.prediction_update(ma, dt)
    kalman_filter.measurement_update( np.matrix([[mp], [ms]]) )
    filtered_pose.append(kalman_filter.get_pose())
    filtered_speed.append(kalman_filter.get_speed())
    covariance.append(kalman_filter.get_covariance())
    
print('* Pose *')
plt.plot(time_line, unit_step, 'g*')
plt.plot(time_line, filtered_pose, 'b*')
```
